**Problem**: String patterns matching

**Input**: A text string t and a patterns string p

**Output**: index of the first occurrence of p appear in t

**Complexity**: O(n<sup>2</sup>)

```kotlin
fun findPattern(t: String, p: String): Int {
    var i = 0;
    while (i < (t.length - p.length)) {
        var j = 0
        while (j < p.length && t[i + j] == p[j]) {
            j++
        }
        if (j == p.length) return i;
        i++
    }
    return -1
}

val text = "The lazy dog jumps over the quick brown fox"
val pattern = "dog"
val index = findPattern(text, pattern)
if (index >= 0) {
    println("Patterns is found at $index, which is ${text.slice(index..(text.length - 1))}")
} else {
    println("Patterns is not found.")
}
```

**Problem**: Sorting

**Input**: A sequence S of elements in an unsorted order

**Output**: A sequence S of elements in a sorted order (e1 < e2, e2 < e3,...) or otherwise

**Time Complexity**: O(n<sup>2</sup>)

```kotlin
fun insertionSort(arr: Array<Int>): Array<Int> {
    for (i in 1 until arr.size) {
        var j = i;
        var count = 0;
        while ((j > 0) && (arr[j - 1] > arr[j])) {
            arr.swap(j - 1, j)
            j--
        }
        count = 0
    }
    return arr
}

fun Array<Int>.swap(index1: Int, index2: Int) {
    var temp = this[index1]
    this[index1] = this[index2]
    this[index2] = temp
}

val arr = arrayOf(5, 7, 2, 4, 8, 9)
insertionSort(arr)
```

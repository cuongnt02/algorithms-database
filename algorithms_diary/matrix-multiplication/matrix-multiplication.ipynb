**Problem**: Matrix multiplication

**Input**: 2 matrices A (dim XxY) and B (dim YxZ)

**Output**: Matrix C (dim XxZ) with each element C[i][j] = C[i] dot C[j] for each i, j is row and column of A and B

**Time Complexity**: O(n<sup>3</sup>)

```kotlin
class Matrix {
    var rows: Int = 0
    var columns: Int = 0
    lateinit var members: Array<Array<Int>>

    fun initialize() {
        members = Array(rows) {
            Array<Int>(columns) { 0 }
        }
    }

    constructor(rows: Int, cols: Int) {
        this.rows = rows
        this.columns = cols
        initialize()
    }

    constructor() {
        initialize()
    }

    operator fun get(index: Int): Array<Int> {
        return members[index]
    }

    operator fun times(other: Matrix): Matrix {
        return matrixMultiplication(this, other)
    }

    companion object {
        fun matrixMultiplication(m1: Matrix, m2: Matrix): Matrix {
            val c = Matrix(m1.rows, m2.columns);
            for (i in 0..m1.rows - 1) {
                for (j in 0..m2.columns - 1) {
                    for (k in 0..m1.columns - 1) {
                         c[i][j] += m1[i][k] * m2[k][j]
                    }
                }
            }
            return c
        }
    }
}

fun readMatrix(m: Matrix) {
    try {
        val (rows, columns) = readln().split(" ").map { it.toInt() }
        m.rows = rows
        m.columns = columns
        var row = 0;
        m.initialize()
        do {
            val members = readln()
            val membersList = members.split(" ").map { it.toInt() }
            if (membersList.size > m.columns) throw IndexOutOfBoundsException()
            membersList.forEachIndexed { index, el ->
                m.members[row][index] = el
            }
            row++
        } while (row < rows)
    } catch (e: IndexOutOfBoundsException) {
        println("Member entered is higher than intended columns")
    }

}

fun printMatrix(m: Matrix) {
    m.members.forEach {
        it.forEach {
            print(" $it")
        }
        println()
    }
}

val A = Matrix()
readMatrix(A)
printMatrix(A)
val B = Matrix()
readMatrix(B)
printMatrix(B)
println("Matrix C = A x B")
val C: Matrix = A*B
printMatrix(C)
```

**Problem**: Power of an integer

**Input**: Integer a and a power n

**Output**: a<sup>n</sup>

```kotlin
// Complexity: O(n)
fun power(a: Int, n: Int): Int {
    var result = 1
    for (i in 0 until n) {
        result *= a
    }
    return result
}

power(2, 2)
power(1, 1000000000)
```

```kotlin
// Complexity: O(logn)
fun power(a: Int, n: Int): Int {
    if (n == 0) return 1
    val x = power(a, n / 2)
    if (n % 2 == 0) return x * x
    else return a * x * x
}

power(2, 3)
power(1, 1000000000)
```

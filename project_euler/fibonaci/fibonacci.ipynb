# Fibonacci

So first thing I wondered is using the brute-force method for finding fibonaci

```kotlin
fun fibonaci(n: Int): Int {
    if (n <= 1) return 1
    return fibonaci(n - 2) + fibonaci(n - 1)
}
```

```kotlin
fibonaci(7)
```

```kotlin
var sum = 0
var i = 0
var fib = fibonaci(i)
while (fib < 4_000_000) {
    if (fib % 2 == 0) {
        sum += fib
    }
    fib = fibonaci(++i)
}
sum
```

Turns out this is the optimal way for this problem. because whether how fast can I find a fibonacci number, there is only a finite set of fibonacci number below 4000000. Which is overall a O(logn) solution.

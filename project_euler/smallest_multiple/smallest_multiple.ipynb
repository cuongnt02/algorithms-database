# Smallest multiple
So, to define the problem:
- A range of number 1-20
- The result number should be evenly divisible to each number in the range

## Brute-force solution
- I have a range of number (1-20), possibly a loop to get each of the number for processing
- I also need to have another loop to loop through all the possible number for processing the result, so in fact we would have 2 loops
- Complexity: O(n^2)

```kotlin
fun smallest_multible(min: Int, max: Int): Int {
    var n = 1
    while (!check_number(n, min, max)) n++
    return n
}

fun check_number(n: Int, min: Int, max: Int): Boolean {
    for (i in min..max) {
        if (n % i != 0) {
            return false
        }
    }
    return true
}

smallest_multible(1, 20)
```

## Slightly faster solution
I can make the program faster by eliminate half of the input number. The main idea is based on the range input. In the rage, there are:
- primes
- multiples of primes

```kotlin
import kotlin.math.sqrt

fun smallest_multible(min: Int, max: Int): Int {
    var primes: List<Int> = emptyList()
    for (i in 2..max) {
        println("Current number: $i")
        println("Check: ${check_prime(i)}")
        if (check_prime(i)) primes += i
    }
    print(primes)
    var n = 1
    while (!check_number(n, primes)) n++
    return n
}

fun check_number(n: Int, primes: List<Int>): Boolean {
    for (i in primes) {
        if (n % i != 0) {
            return false
        }
    }
    var i = 2
    while (i * 2 <= 10) {
        if (n % i != 0) {
            return false
        }
        i++
    }
    return true
}
fun check_prime(n: Int): Boolean {
    for (i in 2..sqrt(n.toDouble()).toInt()) {
        if (n % i == 0) {
            return false
        }
    }
    return true
}

smallest_multible(1, 10)
```

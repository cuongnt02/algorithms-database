# Perfect square
A perfect square or a square number is a square of a positive integer

```python
def perfect_square(n: int) -> int:
    if (n > 0):
        return n*n
    return 0
```

```python
print("Perfect square:")
print(perfect_square(5))
print(perfect_square(10))
```

## Odd square
An odd square, obviously, is a square of a positive odd number

```python
def odd_square(n: int) -> int:
    if n % 2 == 0:
        return 0 # not an odd square
    else:
        return perfect_square(n)
```

```python
print("Odd square:")
print(odd_square(5))
print(odd_square(7))
print(odd_square(10))
```

```python
## Sum of all odd squares
def odd_square_sum(max: int) -> int:
    sum = 0
    for i in range(1, max, 1):
        sum += odd_square(i)
    return sum
```

```python
print(odd_square_sum(254000))
```

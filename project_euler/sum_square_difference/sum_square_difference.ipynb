# Sum Square Difference
This problem to me is kinda straightforward. I loop for one hundred numbers and add it to the sum 

```kotlin
fun square(n: Int): Int = n*n

fun findDifference(n: Int): Int {
    var sumOfSquares = 0
    var squareOfSum = 0
    for (i in 1..n) {
        sumOfSquares += square(i)
        squareOfSum += i
    }
    return squareOfSum*squareOfSum - sumOfSquares
}

findDifference(100)
```

# Optimized version
Turns out I can use the actual sum of a sequence and sum of a square sequence formulas for a O(1) solution.

```kotlin
fun findDifferenceOptimized(n: Int): Int {
    val sum = (n*(n + 1)) / 2
    val squareOfSum = sum*sum
    val sumOfSquares = (n*(n+1)*(2*n+1)) / 6
    return squareOfSum - sumOfSquares
}

findDifferenceOptimized(100)
```
